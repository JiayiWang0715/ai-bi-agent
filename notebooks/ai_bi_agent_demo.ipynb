read data

```python
import pandas as pd

df = pd.read_csv("/content/sample_sales_data.csv")
df
```

basic analytics


```python
summary = {
    "total_sales": df["Sales"].sum(),
    "average_sales": df["Sales"].mean(),
    "best_region_by_sales": df.groupby("Region")["Sales"].sum().idxmax(),
    "lowest_region_by_sales": df.groupby("Region")["Sales"].sum().idxmin(),
    "average_satisfaction": df["Customer_Satisfaction"].mean()
}

summary
```

sammarize based on regions

```python
region_summary = df.groupby("Region").agg({
    "Sales": "sum",
    "Cost": "sum",
    "Customer_Satisfaction": "mean",
    "Follow_Up_Count": "sum"
}).reset_index()

region_summary
```

```python
report = f"""
Business Intelligence Report

Total Sales: {summary['total_sales']}
Average Sales: {summary['average_sales']:.2f}
Best Performing Region: {summary['best_region_by_sales']}
Lowest Performing Region: {summary['lowest_region_by_sales']}
Average Customer Satisfaction: {summary['average_satisfaction']:.2f}

Recommendations:
1. Increase investment in {summary['best_region_by_sales']} due to strong sales performance.
2. Review operations in {summary['lowest_region_by_sales']} to identify growth opportunities.
3. Monitor customer satisfaction trends and follow-up activity for retention improvement.
"""

print(report)
```
